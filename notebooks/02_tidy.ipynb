```python
weather = pd.read_csv('../data/weather.csv')
```

```python
weather
```

```python
weather_melt = weather.melt(id_vars=['id', 'year', 'month', 'element'],
                           var_name='day',
                           value_name='temp')
```

```python
weather_melt.head()
```

```python
weather_tidy = (weather_melt # if not using round brackets, you need \ between space
                .pivot_table(index=['id', 'year', 'month', 'day'], 
                             columns='element', 
                             values='temp')
                .reset_index())
```

```python
weather_tidy
```
