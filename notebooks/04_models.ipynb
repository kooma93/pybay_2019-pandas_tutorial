```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
from sklearn import linear_model
```

```python
lr = linear_model.LinearRegression()
```

```python
lr.fit(X=tips[['total_bill', 'size']], y=tips['tip'])
```

```python
lr.coef_
```

```python
# stats models gives you a summary model but scikit does not
lr.intercept_
```

```python
lr = linear_model.LinearRegression()
lr.fit(X=tips[['total_bill', 'day']], y=tips['tip'])
```

```python
pd.get_dummies(tips) # converts strings into kind of a numerical code
```

```python
tips_dummy = pd.get_dummies(tips, drop_first=True) # converts strings into kind of a numerical code
```

```python
tips_dummy
```

```python
tips_dummy.columns
```

```python
lr = linear_model.LinearRegression()
lr.fit(X=tips_dummy[['total_bill', 'size', 'sex_Female', 'smoker_No', 'day_Fri',
       'day_Sat', 'day_Sun', 'time_Dinner']], y=tips_dummy['tip'])
```

```python
lr.coef_ # if the sex was female, you have increase of 3 cents in tip. If the party shows up on Friday, 16 cents increased compared to Thursday
```
